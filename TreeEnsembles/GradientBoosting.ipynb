# Tree Ensembles: Practical Demonstration (Gradient Boosting)



This Notebook contains the code used in the Medium Article called [Tree Ensembles: Theory and Practice](https://www.google.es) for the Boosting Practical Demonstration

```python
import random
import pandas as pd
import numpy as np

from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree
import matplotlib.pyplot as plt

from synthetic_data import create_data

RANDOM_SEED = 42
NUMBER_OF_TREES = 5
```

## Synthetic Dataset

A small synthetic dataset is created in order to easily show how gradient boosting works. For this notebook, a regression feature is going to be predicted.

```python
dataframe = create_data(categorical=False)
display(dataframe)
```

## Gradient Boosting

Lo primero es hacer los residuals

```python
# First of all, we need to define the first residuals (output_feature - mean(output_feature))
mean_value = dataframe["price"].mean()
dataframe["mean"] = mean_value
dataframe["residuals_0"] = dataframe["price"] - mean_value

display(dataframe)

# Now, let's define the training DataFrame
features = ["square_meters", "has_garage", "has_garden", "rooms"]
dataframe_x = dataframe[features]

# The mean is able to achieve an specific MAE, which will be our baseline to compare the predictions
mae = mean_absolute_error(y_true=dataframe["price"], y_pred=dataframe["mean"])
print(f"MAE of the mean: {mae}")
mae_results = {}
mae_results[f"mean"] = mae

# Here we have the weak learner (Decicion Tree Regressor)
dtc = DecisionTreeRegressor(random_state=RANDOM_SEED, max_depth=2)

# The learning Rate will define how much information each trained tree is going to give to the final model.
learning_rate = 0.3

# Iterating over the trees
for i in range(0, NUMBER_OF_TREES):
    # Training the tree with the residuals 
    dtc.fit(dataframe_x, dataframe[f"residuals_{i}"])
    
    # Getting the residuals predicted
    dataframe[f"residual_predictions_{i}"] = dtc.predict(dataframe_x)
    
    # To the mean value specified in the beginning, we sum up the information that each tree trained before is giving (learning rate * residual_prediction)
    dataframe[f"predictions_{i}"] = mean_value 
    for j in range(0, i + 1):
        dataframe[f"predictions_{i}"] = dataframe[f"predictions_{i}"] + (learning_rate * dataframe[f"residual_predictions_{j}"])
    
    # Finally, we get the residuals again with the new predictions
    dataframe[f"residuals_{i + 1}"] = dataframe["price"] - dataframe[f"predictions_{i}"]
    
    # Plotting the tree
    plt.figure(figsize=(10, 10))
    tree.plot_tree(dtc, feature_names=features)
    plt.show()
    
    # Calculation of the MAE with the last predictions
    mae = mean_absolute_error(y_true=dataframe["price"], y_pred=dataframe[f"predictions_{i}"])
    mae_results[f"tree_{i}"] = mae
    print(f"MAE of tree {i}: {mae}")
    
    # Displaying the final DataFrame with all the information
    display(dataframe)
```

```python
# Showing the accuracy results
pd.DataFrame(mae_results, index=[0]).T.reset_index().rename({"index": "Iteration", 0: "MAE"}, axis="columns")
```
